```python
import arcpy
```

```python
data = r'D:\DataforArcObject\Delhi.gdb\Dataset\Building'
```

```python
arcpy.AddField_management(data,"PolyArea","FLOAT",5)
```

```python
fieldName = 'PolyArea'
expr = '!shape.area!'   # code for shape area calculation
```

```python
arcpy.CalculateField_management(data, fieldName, expr,'PYTHON')
```

```python
rows = arcpy.da.SearchCursor(data,['OBJECTID','PolyArea'])
```

```python
a_row = []
b_row = []
for row in rows:
    print(row[0],' : ',row[1])
    a_row.append(row[0])
    b_row.append(row[1])
del rows
```
